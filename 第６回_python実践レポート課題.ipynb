```python
import ipywidgets as w
from IPython.display import display
import matplotlib.pyplot as plt

辞書 = [['紫陽花','あじさい'], ['竜胆', 'りんどう'], ['譲葉', 'ゆずりは']]
番号 = 0
正解数 = 0

def 結果():
    plt.figure(figsize=(5, 1))
    plt.barh(['OK','NG'],[正解数,len(辞書)-正解数])
    plt.show()

def 漢字表示():
    if 番号 == len(辞書):
      結果() #print('正解数: ' + str(正解数))
      表示.disabled = True
      合ってた.disabled = True
      間違っていた.disabled = True
      return
    漢字.value = 辞書[番号][0]

def fh(btn):
    読み.value = 辞書[番号][1]

def 次():
    global 番号
    番号 += 1
    漢字表示()
    読み.value = '読みは?'

def fm(btn):
    print(辞書[番号][0]+':'+ 辞書[番号][1])
    次()

def fa(btn):
    global 正解数
    正解数 += 1
    次()

# GUI の作成
漢字 = w.Text()
読み = w.Label("読みは？")
表示 = w.Button(description="表示",button_style='success')
合ってた = w.Button(description='合ってた',button_style='info')
間違っていた = w.Button(description='間違っていた',button_style='info')

#イベントハンドラの設定
表示.on_click(fh)
間違っていた.on_click(fm)
合ってた.on_click(fa)

# GUI の表示
display(漢字, 読み, w.HBox(children =[表示,合ってた,間違っていた]))
漢字表示()
```
